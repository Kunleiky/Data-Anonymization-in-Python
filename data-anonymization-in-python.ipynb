## Perform anonymization/obfuscation on the data

### Use of Faker library

```python
# Introduce fake names into the customer name column
dict_names = {name: faker.name( ) for name in edna['Customer Name'].unique()}
edna['Customer Name'] = edna['Customer Name'].map(dict_names)
```

```python
# Check to see fake customer names are applied
edna.head()
```

```python
# Introduce fake emails into the email address column
dict_mails = {mail: faker.email( ) for mail in biz['Email_Address'].unique()}
biz['Email_Address'] = biz['Email_Address'].map(dict_mails)
```

```python
# Check to see that fake email address is applied
biz.head()
```

### Use of Cryptography.fernet - Defining function

```python
# Function to encrypt a value
def encrypt_value(value):
    # Convert the value to bytes (Fernet requires bytes input)
    byte_value = value.encode('utf-8')
    encrypted_value = f.encrypt(byte_value)
    return encrypted_value
```

```python
# Apply encryption to the Customer Name column
edna['Customer Name'] = edna['Customer Name'].apply(encrypt_value)
```

```python
edna.head()
```

```python
 # Apply encryption to the reservation_id column
hospitality['reservation_id'] = hospitality['reservation_id'].apply(encrypt_value)
```

```python
hospitality.head()
```

```python
 # Apply encryption to the Email_Address column
biz['Email_Address'] = biz['Email_Address'].apply(encrypt_value)
```

```python
biz.head()
```

### Use of Cryptography.fernet - Using anonymous (lambda) function

```python
# Apply encryption using a lambda function on the Customer Name column
edna['Customer Name'] = edna['Customer Name'].astype(str).apply(lambda x: f.encrypt(x.encode('utf-8')))
```

```python
edna.head()
```

```python
# Apply encryption using a lambda function on the reservation_id column
hospitality['reservation_id'] = hospitality['reservation_id'].astype(str).apply(lambda x: f.encrypt(x.encode('utf-8')))
```

```python
hospitality.head()
```

```python
# Apply encryption using a lambda function on the Email_Address column
biz['Email_Address']  = biz['Email_Address'] .astype(str).apply(lambda x: f.encrypt(x.encode('utf-8')))
```

```python
biz.head()
```
